```python
# Merging Datas
import os
import pandas as pd


def extract_company_name(file_name):
    return file_name.split('.')[0]


dfs = []


for file_name in os.listdir(path):
    if file_name.endswith('.csv'):
        df = pd.read_csv(os.path.join(path, file_name))
        company_name = extract_company_name(file_name)
        df['company_name'] = company_name
        dfs.append(df)

final_df = pd.concat(dfs, ignore_index=True)
```

```python
final_df.sample(10)
```

```python
# Define a custom function to create custom text for each row
def custom_text(row):
    return f"On {row['Date']}, {row['company_name']} opened trading at ${row['Open']} per share, quickly surged to a high of ${row['High']}, experienced a brief dip to ${row['Low']}, and closed the day at ${row['Close']}. The trading day saw robust activity with a total of {row['Volume']} shares exchanged. Additionally, shareholders received a dividend payment of ${row['Dividends']} per share."

final_df['custom_text'] = final_df.apply(custom_text, axis=1)
```

```python
final_df["custom_text"][0]
```

```python
# a customized funciton to implement the type of investment

def assign_comment(row):
    if row['Close'] < row['Open']:
        if row['Volume'] <= 10000:
            return "Not a good Investment"
        else:
            return "Pretty Good Investment"
    elif row['Close'] > row['Open']:
        if row['Volume'] <= 10000:
            return "Not a good Investment"
        else:
            return "Pretty Good Investment"
    elif row['Close'] == row['Open']:
        if row['Volume'] <= 10000:
            return "Not a good Investment"
        elif row['Volume'] >= 10000 :
            return "Pretty Good Investment"
    

        
final_df['Comment'] = final_df.apply(assign_comment, axis=1)
```

```python

def custom_instruction(row):
    return f"Please evaluate the investment based on the open price, close price and the amount of investment on the volume. Write the ansewer in [Pretty Good Investment or Not a good Investment]"

final_df['instruction'] = final_df.apply(custom_instruction, axis=1)
```

```python
# Select the two columns and rename them
new_df = final_df.loc[:, ['custom_text', 'Comment', 'instruction']].rename(columns={'custom_text': 'input', 'Comment': 'output'})
```
